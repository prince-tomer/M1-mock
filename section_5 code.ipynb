```python
"""
Section 5: Visualization + Probability – 10 Marks
Tasks:
1.	Create:
○	1 Bar chart
○	1 Line chart
2.	Probability Question:
○	If 30 out of 100 students passed, what is probability of passing?
○	Explain in context of dataset
"""
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Customers1 - Sheet1.csv")

# Handle missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())

# 🔹 Bar Chart: Count of Customers by Region
region_count = df['Region'].value_counts()

plt.figure()
region_count.plot(kind='bar')
plt.title("Number of Customers by Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()


# 🔹 Line Chart: Age Distribution (sorted)
sorted_df = df.sort_values(by='Age')

plt.figure()
plt.plot(sorted_df['Age'].values)
plt.title("Customer Age Trend")
plt.xlabel("Customer Index")
plt.ylabel("Age")
plt.show()
```
