```python
"""Section 4: 
Data Handling + Analysis – 15 Marks
Dataset:
Use any CSV (e.g., student performance / sales / HR data)
Tasks:
1.	Load dataset using Pandas
2.	Perform:
○	Filtering
○	Grouping
○	Sorting
3.	Generate 2 insights
👉 Example:
●	“Average score by subject”
●	“Top performing category”
________________________________________
"""
# ✅ Import Library
import pandas as pd

# ✅ 1. Load Dataset
df = pd.read_csv("Customers1 - Sheet1.csv")

print("🔹 Original Dataset:")
print(df)
print("\n-----------------------------\n")

# ✅ Handle Missing Values (optional but good practice)
df['Age'] = df['Age'].fillna(df['Age'].mean())

# ✅ 2. Data Operations

# 🔹 Filtering: Customers with Age > 30
filtered = df[df['Age'] > 30]
print("🔹 Filtered Data (Age > 30):")
print(filtered)
print("\n-----------------------------\n")

# 🔹 Grouping: Average Age by Region
grouped = df.groupby('Region')['Age'].mean()
print("🔹 Average Age by Region:")
print(grouped)
print("\n-----------------------------\n")

# 🔹 Sorting: Sort by Age (Descending)
sorted_df = df.sort_values(by='Age', ascending=False)
print("🔹 Sorted Data (Age Descending):")
print(sorted_df)
print("\n-----------------------------\n")

# ✅ 3. Insights

print("🔹 Insights:")

# Insight 1: Region with highest average age
max_region = grouped.idxmax()
max_age = grouped.max()
print(f"1. Region with highest average age: {max_region} ({max_age:.2f})")

# Insight 2: Top 2 oldest customers
top_customers = sorted_df.head(2)
print("\n2. Top 2 oldest customers:")
print(top_customers[['Name', 'Age', 'Region']])
```
